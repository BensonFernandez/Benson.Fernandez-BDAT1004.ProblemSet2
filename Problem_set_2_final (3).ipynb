As shown above, if the file cannot be found by the interpreter or if it cannot be read<br>
as a text file, an exception will be raised. Modify function fileLength() so that a<br>
friendly message is printed instead:<br>
 fileLength('midterm.py')<br>
358<br>
 fileLength('idterm.py')<br>

```python
import os;

def fileLength(filename):
    file_exists = False
    if filename in os.listdir():
        file_exists = True

    if file_exists:
        with open(filename, 'r') as infile:
            content = infile.read()
            return "File {} length = {}".format(filename, len(content))
    else:
        return "File {} not found. Please enter the correct file name".format(filename) # added code to display a statement for absent files


print(fileLength('idterm.py'))       
print(fileLength('midterm.py'))
```

Write a class named Marsupial that can be used as shown below:<br>

m.put_in_pouch('doll')<br>

m.pouch_contents()<br>
['doll', 'firetruck', 'kitten']<br>
Now write a class named Kangaroo as a subclass of Marsupial that inherits all the<br>
attributes of Marsupial and also:<br>
a. extends the Marsupial __init__ constructor to take, as input, the<br>
coordinates x and y of the Kangaroo object,<br>
b. supports method jump that takes number values dx and dy as input and<br>
moves the kangaroo by dx units along the x-axis and by dy units along the y-axis, and<br>
c. overloads the __str__ operator so it behaves as shown below.

```python
class Marsupial:
    def __init__(self):
        self.pouch = []  # empty list pouch is created for Marsupial

    def put_in_pouch(self, item):
        self.pouch.append(item)  # To add 'item' to the list pouch

    def pouch_contents(self):
        return self.pouch  # returns the value of list pouch.


class Kangaroo(Marsupial):
    def __init__(self, x, y):
        super().__init__()  #  To call init method of Marsupial
        self.x = x  #  'x' coordinate of Kangaroo
        self.y = y  #  'y' coordinate of Kangaroo

    def jump(self, dx, dy):
        self.x += dx  # dx is added to the 'x' coordinate of the Kangaroo
        self.y += dy  # dy is added to the 'y' coordinate of the Kangaroo 

    def __str__(self):
        return "I am a Kangaroo located at coordinates ({}, {})".format(self.x, self.y)
    

m = Marsupial()
m.put_in_pouch('doll')
m.put_in_pouch('firetruck')
m.put_in_pouch('kitten')
print(m.pouch_contents()) 

k = Kangaroo(0, 0)
print(k) 
k.put_in_pouch('doll')
k.put_in_pouch('firetruck')
k.put_in_pouch('kitten')
print(k.pouch_contents())  
k.jump(1, 0)
k.jump(1, 0)
k.jump(1, 0)
print(k)  
```

# Question 4
Write function collatz() that takes a positive integer x as input and prints the<br>
Collatz sequence starting at x. A Collatz sequence is obtained by repeatedly applying<br>
this rule to the previous number x in the sequence:<br>
x = {<br>
𝑥/2 𝑖𝑓 𝑥 𝑖𝑠 𝑒𝑣𝑒𝑛<br>
3𝑥 + 1 𝑖𝑓 𝑥 𝑖𝑠 𝑜𝑑𝑑<br>
Your function should stop when the sequence gets to number 1. Your<br>
implementation must be recursive, without any loops.

```python
def collatz(x):
    sequence = [x]  # list is initialized to hold the binary digits
    
    for _ in range(1000):  # used to iterate a maximum of 1000 times to avoid infinite loops
        if x == 1:
            break
        
        if x % 2 == 0: 
            x = x // 2 
        else:  # If x is odd
            x = 3 * x + 1 
        
        sequence.append(x)  # to add the new value of x to the sequence list
    
    print(*sequence, sep='\n') # \n is used to push the next value of iteration to next line

collatz(12)
```

# Question 5
Write a recursive method binary() that takes a non-negative<br>
integer n and prints the binary representation of integer n.<br>
binary(0)<br>
0<br>
binary(1)<br>
1<br>
binary(3)<br>
11<br>
binary(9)<br>
1001<br>

```python
def binary(n):
    bits = []  #  a list to store the binary bits
    
    while n > 0:  
        bits.append(n % 2)  
        n = n // 2  
    
    if len(bits) == 0:
        bits.append(0)
    
    print(*reversed(bits), sep='', end='')  # Print the reversed list of bits without separators, ending with an empty string

binary(111)  
```

# Question 6
Implement a class named HeadingParser that can be used to parse an HTML document, and retrieve and print all the headings in the document. You should implement your class as a subclass of HTMLParser, defined in Standard Library module html.parser. When fed a string containing HTML code, your class should print the headings, one per line and in the order in which they appear in the
document. Each heading should be indented as follows: an h1 heading should have indentation 0, and h2 heading should have indentation 1, etc. Test your implementation using w3c.html.

```python
from html.parser import HTMLParser

class HeadingParser(HTMLParser):
    def __init__(self):
        super().__init__()
        self.headings = []
        self.in_body = False
        self.indentation_levels = {
            'h1': 0,
            'h2': 1,
            'h3': 2,
            'h4': 3,
            'h5': 4,
            'h6': 5
        }# indentation levels for different heading tags 

    def handle_starttag(self, tag, attrs):
        if tag.lower() == 'body':
            self.in_body = True

        tag = tag.lower()
        if tag in self.indentation_levels:
            self.indentation = self.indentation_levels[tag]

    def handle_endtag(self, tag):
        if tag.lower() == 'body':
            self.in_body = False

    def handle_data(self, data):
        if self.in_body:
            content = data.strip()# Strip leading and trailing whitespace from the data
            if content:
                indented_content = ' ' * self.indentation + content
                self.headings.append(indented_content)# Add the content to the list of headings

    def print_headings(self):
        print(*self.headings, sep='\n')

with open('w3c.html') as infile:
    content = infile.read()
    infile.close()

hp = HeadingParser()
hp.feed(content)
hp.print_headings()
```

# Question 7
Implement recursive function webdir() that takes as input: a URL (as a string) and non-negative integers depth and indent. Your function should visit every web page reachable from the starting URL web page in depth clicks or less, and print each web page's URL. As shown below, indentation, specified by indent, should be used to indicate the depth of a URL.<br>

webdir('http://reed.cs.depaul.edu/lperkovic/csc242/test1.html', 2, 0)<br>
http://reed.cs.depaul.edu/lperkovic/csc242/test1.html<br>
_____http://reed.cs.depaul.edu/lperkovic/csc242/test2.html<br>
__________http://reed.cs.depaul.edu/lperkovic/csc242/test4.html<br>
_____http://reed.cs.depaul.edu/lperkovic/csc242/test3.html<br>
__________http://reed.cs.depaul.edu/lperkovic/csc242/test4.html<br>

```python
import requests
import re

def webdir(url, depth, indent):
    print(' ' * indent + url)

    if depth == 0:
        return

    response = requests.get(url) # request to get content of URL
    if response.ok:
        content = response.text

        links = extract_links(content) # Extract links from the content

        for link in links:
            webdir(link, depth - 1, indent + 4)
    else:
        print('Cannot access URL:', url)
        print('Error:', response.status_code)

def extract_links(content):
    url_pattern = r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+'
    links = re.findall(url_pattern, content) # Find all URLs in the content using regex pattern
    return links

webdir('https://www.ruinedking.com/', 2, 0)
```
